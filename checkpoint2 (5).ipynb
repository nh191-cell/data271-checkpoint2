## Data Sources

The first dataset used in this project comes from the U.S. Environmental Protection Agency (EPA):

https://www.epa.gov/climate-indicators/climate-change-indicators-sea-level

The data tracks global sea level changes over time and is used to study climate change trends. It includes variables such as year and sea level measurements.

Some values were missing (e.g., NOAA Adjusted Sea Level), and the dataset required cleaning and converting data types to numeric for analysis.

Additionally, a second dataset was used from:

https://raw.githubusercontent.com/datasets/global-temp/master/data/monthly.csv

This dataset tracks global temperature anomalies over time and is used to study climate change. The data spans from the mid-1800s to the present and includes variables such as Date and Mean temperature.

Since the data is recorded monthly, it was converted into yearly averages for analysis. This dataset will be combined with the sea level dataset using the Year variable to explore the relationship between temperature and sea level rise.

## Load Sea Level Data

This section loads the sea level dataset and prepares it for analysis.

```python
url = "https://raw.githubusercontent.com/datasets/sea-level-rise/master/data/epa-sea-level.csv"
df = pd.read_csv(url)
df.head()
```

## Clean Sea Level Data

This section cleans the dataset by renaming columns and converting values to numeric for analysis.

```python
df.info()
df.describe()
```

## Load Temperature Data

This section loads the global temperature dataset for analysis.

```python
temp = pd.read_csv("https://raw.githubusercontent.com/datasets/global-temp/master/data/monthly.csv")
temp.head()
```

## Clean Temperature Data

This section prepares the temperature dataset by extracting the year and calculating yearly average temperature values.

```python
temp["Year"] = temp["Year"].astype(str).str[:4]
temp["Year"] = pd.to_numeric(temp["Year"], errors="coerce")

temp_yearly = temp.groupby("Year")["Mean"].mean().reset_index()

temp_yearly.head()
```

```python
merged = pd.merge(df, temp_yearly, on="Year", how="inner")
merged.head()
```
